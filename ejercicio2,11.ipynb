```
def factorial(n):
  f=1.0
  for j in range (1, n+1):
    f *=j
  return f
```

```

def Binomial(n,k):
  m=n-k
  o=factorial(int(m))
  s=factorial(int(k))
  t=factorial(int(n))
    
  h=t/(o*s)
  p=int(h)
  return p
```
