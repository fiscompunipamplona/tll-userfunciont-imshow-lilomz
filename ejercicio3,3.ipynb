```
import numpy as np
from numpy import loadtxt, empty
from pylab import imshow, gray, show, jet, hot, hsv, xlim, ylim
import math
from math import sqrt,sin,pi

```

```
data=loadtxt("stm.txt",float)

imshow(data, origin="lower")
gray()
show()

imshow(data, origin="lower")
xlim(0,300)
ylim(0,300)
hsv()
show()

imshow(data, origin="lower", extent=[0,10,0,5])
jet()
show()

imshow(data, origin="lower", extent=[0,10,0,5], aspect=2.0)
hot()
show()

```
